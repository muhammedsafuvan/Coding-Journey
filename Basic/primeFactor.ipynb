```python
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
n = 14997430413624714359409666680051077987067321755389882420186601162721136432857785455555834921723408822394276417807670465730154635705807256401852896333103255651692991996926647870684397859884099079116148988105414536374565537412149658635826927826197166412472368080654743323404117014615118698981754247988242523495672794755417015794585469422695983928999736680513369315369660015832564296216765340286920082930856335672690891274415924071513886916722661666048622084196367834359134898920056623728535931606703976462511384760821492392448304916704430673337129470659457651752686871152090578554014878864880957019376221107662618251427
print(prime_factors(n))
```

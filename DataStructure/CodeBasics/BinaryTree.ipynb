```python
class BinaryTree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
    def add_child(self, data):
        if self.data == data:
            return
        
        if data < self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left = BinaryTree(data)
        else:
            if self.right:
                self.right.add_child(data)
            else:
                self.right = BinaryTree(data)
                
    def in_order(self):
        elements = []
        if self.left:
            elements += self.left.in_order()
        
        elements.append(self.data)
        
        if self.right:
            elements += self.right.in_order()
            
        return elements
    
    def pre_order(self):
        elements = []
        
        elements.append(self.data)
        
        if self.left:
            elements += self.left.in_order()
        
        
        if self.right:
            elements += self.right.in_order()
            
        return elements
    
    def post_order(self):
        elements = []
        
        
        if self.left:
            elements += self.left.in_order()
        
        
        if self.right:
            elements += self.right.in_order()
            
        elements.append(self.data)
            
        return elements
    
    def search(self, val):
        if self.data == val:
            return True
        
        if val < self.data:
            if self.left:
                return self.left.search(val)
            else:
                return False
            
        if val > self.data:
            if self.right:
                return self.right.search(val)
            else:
                return False
            
    def find_min(self):
        if not self.left:
            return self.data
        else:
            return self.left.find_min()
        
    def find_max(self):
        if not self.right:
            return self.data
        else:
            return self.right.find_max()
        
    def find_sum(self):
        ans = 0
        
        if self.left:
            ans += self.left.find_sum()
        ans += self.data
        
        if self.right:
            ans += self.right.find_sum()
        
        return ans
            
        
    def delete_using_right_sub_tree(self, val):
        
        
        if self.data > val:
            if self.left:
                self.left = self.left.delete_using_right_sub_tree(val)
                
        elif self.data < val:
            if self.right:
                self.right = self.right.delete_using_right_sub_tree(val)
           
        else:
            
            if self.left is None and self.right is None:
                return None
            
            if self.left is None:
                return self.right
            
            if self.right is None:
                return self.left
            
            min_val = self.right.find_min()
            self.data = min_val
            self.right = self.right.delete_using_right_sub_tree(min_val)
        
        return self
    
    def delete_using_left_sub_tree(self, val):
        
        
        if self.data > val:
            if self.left:
                self.left = self.left.delete_using_right_sub_tree(val)
                
        elif self.data < val:
            if self.right:
                self.right = self.right.delete_using_right_sub_tree(val)
           
        else:
            
            if self.left is None and self.right is None:
                return None
            
            if self.left is None:
                return self.right
            
            if self.right is None:
                return self.left
            
            max_val = self.right.find_max()
            self.data = max_val
            self.right = self.right.delete_using_right_sub_tree(min_val)
        
        return self
    
    
    def bfs(self, root):
        
        if not root:
            return
        
        queue = []
        
        queue.append(root)
        
        
        while queue:
            
            print(queue[0].data)
            node = queue.pop(0)
            
            if node.left:
                queue.append(node.left)
                
            if node.right:
                queue.append(node.right)
    
    
    
                

def build_binary_tree(elements):
    root = BinaryTree(elements[0])
    
    for i in range(1, len(elements)):
        root.add_child(elements[i])
        
    return root
        

# elements = [17, 4, 1, 20, 9, 46,23, 18, 34, 18, 4, 1]
elements = [3, 1, 2, 4, 5]
   
tree = build_binary_tree(elements)
# print(tree.pre_order())
# print(tree.post_order())
# print(tree.in_order())
print(tree.search(3))
# print(tree.find_min())
# print(tree.find_max())
print(tree.find_sum())
# tree.delete_using_right_sub_tree(2)
# tree.delete_using_left_sub_tree(2)
# tree.delete_using_left_sub_tree(4)
# tree.delete_using_left_sub_tree(5)
print(tree.pre_order())

root = BinaryTree(1)
root.left = BinaryTree(2)
root.right = BinaryTree(3)
root.left.left = BinaryTree(4)
root.left.right = BinaryTree(5)
tree.bfs(root)
```

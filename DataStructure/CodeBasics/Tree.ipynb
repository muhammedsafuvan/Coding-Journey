```python
class TreeNode:
    def __init__(self,name, designation):
        self.name = name
        self.designation = designation
        self.children = []
        self.parent = None
        
    def add_child(self, child):
        child.parent = self
        self.children.append(child)
        
    def print_tree(self, choice):
        level = self.level()
        
        if choice == "both":
        
            print("\t"*level+"|___",f"{self.name}({self.designation})")
            if self.children:
                for child in self.children:
                    child.print_tree(choice)
                    
        elif choice == "name":
            print("\t"*level+"|___",f"{self.name}")
            if self.children:
                for child in self.children:
                    child.print_tree(choice)
                    
        elif choice == "designation":
            print("\t"*level+"|___",f"{self.designation}")
            if self.children:
                for child in self.children:
                    child.print_tree(choice)
        else:
            print("Invalid Choice.")
                    
        
                
    def level(self):
        l = 0
        p = self.parent
        while p:
            p = p.parent
            l += 1
        return l
    
def build_tree():
    root = TreeNode("Nil", "CEO")
    
    cto = TreeNode("Nil", "CTO")
    
    hr = TreeNode("Gels", "HR Head")
    hr.add_child(TreeNode("Peter","Rec manager"))
    hr.add_child(TreeNode("Waqas","Policy manager"))
    
    vishwa = TreeNode("Vishwa", "Dep Head")
    vishwa.add_child(TreeNode("Dhaval", "Cloud Manager"))
    vishwa.add_child(TreeNode("Abhi", "App Manager"))
    
    cto.add_child(vishwa)
    cto.add_child(TreeNode("Amir", "App Head"))
    
    root.add_child(cto)
    root.add_child(hr)
    
    
    
#     mobile = TreeNode("Mobile")
#     mobile.add_child(TreeNode("iPhone"))
#     mobile.add_child(TreeNode("Samsung"))
#     mobile.add_child(TreeNode("Vivo"))
    
#     lap = TreeNode("Lap")
#     lap.add_child(TreeNode("Mac"))
#     lap.add_child(TreeNode("Dell"))
#     lap.add_child(TreeNode("HP"))
    
#     tv = TreeNode("TV")
#     tv.add_child(TreeNode("LG"))

#     root.add_child(mobile)
#     root.add_child(lap)
#     root.add_child(tv)
    
    return root
        
        
if __name__ == "__main__":
    root = build_tree()
    root.print_tree("name")
    root.print_tree("both")
    root.print_tree("designation")
    print(root.level())
```
